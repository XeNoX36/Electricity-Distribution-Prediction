```python
data = data.dropna(how="all")
```

```python
# filling-in all the missing values
data[["hour", "dayofweek", "month", "year", "dayofyear"]] = data[["hour", "dayofweek", "month", "year", "dayofyear"]].ffill()
```

```python
data[["Temperature", "Humidity"]] = data[["Temperature", "Humidity"]].bfill()
```

```python
data["Demand"] = data["Demand"].interpolate(method="time")
data["Demand"]
```

```python
data["Demand"].isnull().sum()
```

```python
# creating yearly-quarter column
data.insert(5, "quarter", data.index.quarter)
data
```

```python
# converting the dates to integers
data[["hour", "dayofweek", "month", "year", "dayofyear"]] = data[["hour", "dayofweek", "month", "year", "dayofyear"]].astype(int)
```

```python
# creating 'week of year' column
data.insert(5, "weekofyear", data.index.isocalendar().week.astype(int))
```

```python
data.tail(10)
```

```python
# creating 'weekend' column
data.insert(5, "is_weekend", data.index.dayofweek.isin([5,6]))
```

```python
data["is_weekend"] = data["is_weekend"].astype(int)
```

```python
data.head()
```

```python
# creating a 'holiday' column
import holidays
```

```python
data["holidays"] = holidays.IN(years=data.year)
data
```

```python
data.holidays.value_counts()
```

```python
data = data.drop("holidays", axis=1)
```

```python
# creating a lag column for demand from th same hour yesterday
data["Demand_lag_24hr"] = data["Demand"].shift(24)
data.head(30)
```

```python
# creating a lag column for demand from th same hour last week
data["Demand_lag_168hr"] = data["Demand"].shift(168)  # 168hrs - weekly shift
```

```python
data.iloc[168:200]
```

```python
# creating a 'rolling mean' column
data["Demand_rolling_mean_24hr"] = data["Demand"].rolling(window=24).mean()
data
```

```python
# creating a 'rolling standard deviation' column
data["Demand_rolling_std_24hr"] = data["Demand"].rolling(window=24).std()
data.head(27)
```

```python
# dropping all na rows
data = data.dropna()
data
```

Creating the model

```python
Y = data.Demand  # target 
```

```python
X = data.drop("Demand", axis=1)  # feature
```

```python
# splitting data into training and testing data
# using 80% of data as training data
X_train = X.loc[: "2023-12-31"] 
Y_train = Y.loc[: "2023-12-31"]
Y_train
```

```python
# using 20% of data as testing data
X_test = X.loc["2024-01-01" :] 
Y_test = Y.loc["2024-01-01" :]
Y_test
```

```python
from xgboost import XGBRegressor
import scipy
```

```python
# to evealuate the performance of the model
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
```

```python
# initializing and training the model
model_xgb = XGBRegressor(n_estimators=1000,
                         early_stopping_rounds=50,
                         learning_rate=0.01,
                         objective="reg:squarederror")
```

```python
model_xgb.fit(X_train, Y_train, eval_set=[(X_train, Y_train), (X_test, Y_test)], verbose=False)
```

```python
prediction_xgb = model_xgb.predict(X_test)
```

```python
rmse_xgb = np.sqrt(mean_squared_error(Y_test, predictions_xgb))
mae_xgb = mean_absolute_error(Y_test, predictions_xgb)
print("XGBoost RMSE:" rmse_xgb)
print("XGBoost MAE:" mae_xgb)
```

```python
# Visualize the Predictions

plt.figure(figsize= (15,6))
plt.plot(Y_test.index, Y_test, label="Actual Demand", color="Blue")
plt.plot(Y_test.index, prediction_xgb, label="Predicted Demand", color="Red", linestyles="--")
plt.title("XGBoost Electricity Demand Prediction")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.legend();
```
